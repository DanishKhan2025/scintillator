```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
DATA_DIR = "../data/april_16_3am/"
FILE_NAME = "F1--amp_1--00000.csv"
amplitudes = pd.read_csv(DATA_DIR + FILE_NAME).to_numpy()
print(amplitudes)
```

```python
event_numbers   = np.array([i[0] for i in amplitudes])
peak_sizes      = np.array([i[1] for i in amplitudes])

#printing stuff
print(f"Mean peak amplitude (mV): {np.mean(peak_sizes)}")
print(f"Error in mean (mV): {np.std(peak_sizes)}")

fig, ax = plt.subplots(2)
fig.set_size_inches (10, 10)
fig.set_dpi         (150)
ax[0].plot             (event_numbers, peak_sizes, color="#5BCEFA")
ax[0].set_xlabel       ("Event number")
ax[0].set_ylabel       ("Peak amplitude (V)")
ax[0].set_title        ("Peak amplitudes")

ax[1].hist             ((peak_sizes), bins=500, color=["#F5A9B8"])
ax[1].set_xlabel       ("particle energies (oscilloscope readout (V))")
ax[1].set_ylabel       ("number of counts")
ax[1].set_title        ("distribution of particle energies")
```
