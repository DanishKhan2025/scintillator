```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
DATA_DIR = "../data/april_16_2am/"
FILE_NAME = "F1--amp_1--00000.csv"
amplitudes = pd.read_csv(DATA_DIR + FILE_NAME).to_numpy()
print(amplitudes)
```

```python
event_numbers   = np.array([i[0] for i in amplitudes])
peak_sizes      = np.array([i[1] for i in amplitudes])

#printing stuff
print(f"Mean peak amplitude (mV): {np.mean(peak_sizes)}")
print(f"Error in mean (mV): {np.std(peak_sizes)}")

fig, ax = plt.subplots()
fig.set_size_inches (10, 5)
fig.set_dpi         (200)
ax.plot             (event_numbers, peak_sizes)
ax.set_xlabel       ("Event number")
ax.set_ylabel       ("Peak amplitude (mV)")
ax.set_title        ("Peak amplitudes")


fig, ax = plt.subplots()
fig.set_size_inches (10, 5)
fig.set_dpi         (200)
ax.hist             (peak_sizes, bins=100)
ax.set_xlabel       ("particle energies (oscilloscope readout)")
ax.set_ylabel       ("number of counts")
ax.set_title        ("distribution of particle energies")
```
